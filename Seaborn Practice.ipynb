```python
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt
import matplotlib as mpl # this actually imports matplotlib
import matplotlib.cm as cm #allows us easy access to colormaps

import seaborn as sns
sns.set(color_codes=True)
%matplotlib inline
mpl.rcParams['figure.figsize'] = (8, 5)
```

```python
df = pd.read_csv('dataset/ign.csv')
df.head()
df.dtypes
```

```python
with sns.axes_style('ticks'):
    sns.distplot(df['score'], bins = 10, kde = False)
    #sns.distplot(df['score'], hist = False)
    #sns.kdeplot(df['score'], shade = True, bw = .8)
    plt.xlabel('Score')
    plt.ylabel('Counts')
    plt.title('Game Review Score')
    sns.despine(trim = True, offset = 10, left = False, bottom = False)
```

```python
with sns.axes_style('white'):
    sns.jointplot(x = 'score', y = 'score', data = df, kind = 'hex', color = 'k')
```

```python
iris = sns.load_dataset("iris")
with sns.axes_style('darkgrid'):
    sns.pairplot(iris)
```

```python
with sns.axes_style():
    sns.regplot(data = iris, x = 'sepal_length', y = 'sepal_width')
```

```python
with sns.axes_style():
    sns.lmplot(data = df, x = 'release_month', y = 'score', x_estimator= np.mean)
    #sns.lmplot(data = df, x = 'release_month', y = 'score', x_jitter = .05)
```

```python
with sns.axes_style():
    sns.lmplot(x = 'sepal_length', y = 'sepal_width', data = iris, 
              ci = None, scatter_kws={"s": 20}, order = 1, robust = True)
```

```python
with sns.axes_style():
    sns.residplot(x = 'sepal_length', y = 'sepal_width', data = iris, 
              scatter_kws={"s": 20})
```

```python

with sns.axes_style():

    sns.lmplot(x = 'sepal_length', y = 'sepal_width', data = iris, hue = 'species', markers= ['o', 'x', 'o'], size= 6,
              aspect = .8)
    #sns.lmplot(x = 'sepal_length', y = 'sepal_width', data = iris, col = 'species')
    
```
